```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, matplotlib as mpl
import mplhep as hep
hep.style.use("CMS")

plt.rcParams['figure.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
plt.rcParams['savefig.bbox']='tight'
import uproot


plt.rcParams["figure.figsize"] = (10, 8)
```

```python
def Landau(x, normalization,location,stdev):
    #print(type(x))
    u=(x-location)*3.591/stdev/2.355
    renormalization = 1.64872*normalization
    return renormalization * np.exp(-u/2 - np.exp(-u)/2)

import uproot as ur
events = ur.open(f'dd4hep/zdc_staggered_h4_sim_mu-_50GeV_theta_deg.edm4hep.root:events')
#events = ur.open(f'insert_sim_mu-_18GeV_theta_178-178deg.edm4hep.root:events')
arrays_sim = events.arrays()
#print(arrays_sim.fields)
all_energies=[]
for event in range(len(arrays_sim)):
    all_energies += list(arrays_sim['ZDCHits.energy'][event,:])
h,x=np.histogram(np.array(all_energies)*1000, bins=100, range=(0, 1))

bc=(x[1:]+x[:-1])/2
p0=[100,1000*np.mean(all_energies), -1000*np.std(all_energies)*.8]


slc=(bc>.35)&(bc<0.7) 
from scipy.optimize import curve_fit
coeff, var_matrix = curve_fit(Landau, bc[slc], h[slc], p0=p0,
                         sigma=np.sqrt(h[slc]))
plt.subplots(1,1)
plt.errorbar(bc, h, np.sqrt(h), marker='o', ls='')

plt.xlabel("hit energy [MeV]")
plt.title("$\mu^-$ simulations")
x=np.linspace(bc[0], bc[-1], 100)
plt.plot(x, Landau(x,*coeff))
print(f"mu={coeff[1]:.3}+-{np.sqrt(var_matrix[1][1]):.3}  MeV")
plt.show()
```
