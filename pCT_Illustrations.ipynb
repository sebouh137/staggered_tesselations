```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt, matplotlib as mpl
import random

import mplhep as hep

hep.style.use("CMS")

plt.rcParams['figure.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
plt.rcParams['savefig.bbox']='tight'
```

```python
def draw_hexagon(center, sidelength, **arg):
    phi=np.linspace(-np.pi, np.pi, 7)
    x=center[0]+sidelength*np.cos(phi)
    y=center[1]+sidelength*np.sin(phi)
    plt.plot(x,y, **arg)
```

```python
h,w=9.6, 9.6
plt.plot([w/2,-w/2,-w/2, w/2,w/2],[h/2,h/2,-h/2,-h/2, h/2])
colors="krgb"
offsets=(0,-np.sqrt(3)/4),(0, np.sqrt(3)/4),(3/4, 0),(-3/4, 0)
for k, color in enumerate(colors):
    arg=dict(color=color)
    sidelength=1.25
    for i in range(-2, 3):
        for j in range(-4,5):
            if (j +i)%2:
                continue
            center=sidelength*(i*3/2+offsets[k][0]), sidelength*(j*np.sqrt(3)/2+offsets[k][1])
            draw_hexagon(center, sidelength,**arg)
plt.show()


    
```
