```python
def rowcol(X,Y):
    offsetX=0
    offsetY=0
    sidelength=1
    a=np.mod((Y-offsetY)/(np.sqrt(3)*sidelength),1)
    b=np.mod((X-offsetX)/(3*sidelength),1)
    c= np.floor((X-offsetX)/(3*sidelength/2))+\
            (b<1/2)*(-abs(a-.5)<(b-.5)*3)+(b>1/2)*(abs(a-.5)-.5<(b-1)*3)
    r=np.floor((Y-offsetY)/(np.sqrt(3)*sidelength/2))
    r-=(r+c)%2
    return r,c


```

```python
def position(row,col):
    _sideLength=1
    _offsetX=0
    _offsetY=0
    x=col*1.5*_sideLength+_offsetX+_sideLength/2
    y=row*np.sqrt(3)/2*_sideLength+ _offsetY+_sideLength*np.sqrt(3)/2
    return x,y
```

```python
import numpy as np
from matplotlib import pyplot as plt 
import matplotlib.cm as cm 
n = 256 
x = np.linspace(-5., 5., n) 
y = np.linspace(-5., 5., n) 
X, Y = np.meshgrid(x, y) 

rc = rowcol(X,Y)

phi=np.linspace(-np.pi, np.pi, 7)
plt.plot(np.cos(phi)+1/2, np.sin(phi)+np.sqrt(3))
plt.plot(np.cos(phi)+1/2, np.sin(phi))
plt.plot(np.cos(phi)+2, np.sin(phi)+np.sqrt(3)/2)
plt.plot([0, 0, 3, 3,0], [-np.sqrt(3)/2, np.sqrt(3)*3/2,np.sqrt(3)*3/2, -np.sqrt(3)/2,-np.sqrt(3)/2])

plt.plot([position(0,0)[0]], [position(0,0)[1]], marker='o')
plt.plot([position(1,1)[0]], [position(1,1)[1]], marker='o')


#plt.pcolormesh(X, Y, Z, cmap = cm.gray) 

Z=rc[1]+.3*rc[0]
plt.pcolormesh(X, Y, Z, cmap = cm.gray)#, alpha=0.2) 
plt.show()
```
