```python
from scipy.optimize import curve_fit

def gauss(x, A,mu, sigma):
    return A * np.exp(-(x-mu)**2/(2*sigma**2))
#do just the calculation of the resolution
def res_calc(a, fnc=gauss, bins=None, nsigma=2, quantile=None):
    a=np.array(a)
    mean=np.mean(a)
    sigma=np.std(a)
    if bins is None :
        if quantile==None:
            bins = np.linspace(mean-3*sigma,mean+3*sigma, 20)
        else:
            bins = np.linspace(np.quantile(a, quantile),np.quantile(a, 1-quantile), 20)
    h,x=np.histogram(a, bins=bins)
    bc=(x[1:]+x[:-1])/2
    a_in_window=a[(a>=x[0])&(a<=x[-1])]
    p0=[10,np.mean(a_in_window), np.std(a_in_window)*.9]
    slc=np.abs(bc-np.mean(a_in_window))<nsigma*np.std(a_in_window)
    try:
        coeff, var_matrix = curve_fit(fnc, bc[slc], h[slc], p0=p0,
                                 sigma=np.sqrt(h[slc]))
        return abs(coeff[2]),np.sqrt(var_matrix[2][2]),coeff[1], np.sqrt(var_matrix[1][1])
    except:
        return 0, 0,0,0
```

```python
def res_plot(df, label=f"ZDC (64 Fe) with HEXPLIT, H4 staggering", fit=True, nEnergyBins=15,shift=0, abs=False,alt_residual=True, 
            use_theta_directly=False,
            direction='r', rw="_rw", w0_params=(5.0,0.65,0.31), quantile=None, **arg):
    energy_bins=np.exp(np.linspace(np.log(10), np.log(300), nEnergyBins))
    
    bcs=[]
    sigmas=[]
    dsigmas=[]
    mus=[]
    dmus=[]
    
    for j in range(len(energy_bins)-1):
        dfq=df.query(f"mc_pz>{energy_bins[j]} and mc_pz<{energy_bins[j+1]} and theta_mrad<4")
        if len(dfq)==0:
            continue
        l=np.log((energy_bins[j]+energy_bins[j+1])/2/50)
        #w0=5.0+0.65*l+0.31*l*l
        w0=w0_params[0]+w0_params[1]*l+w0_params[2]*l*l
        w0str=f'{int(w0*4)/4}'.replace('.', 'pt')
        #x=dfq[f'x_truth_face']
        if not alt_residual:
            dr=dfq[f'dr{rw}_w0_{w0str}']
        elif not use_theta_directly:
            dx=dfq.eval(f'dx{rw}_w0_{w0str}/z_truth_w0_{w0str}*{z_front_face*1000}')
            dy=dfq.eval(f'dy{rw}_w0_{w0str}/z_truth_w0_{w0str}*{z_front_face*1000}')
            dr=(dx*dfq.eval("x_truth_face")+dy*dfq.eval("y_truth_face"))/dfq.eval("sqrt(x_truth_face**2+y_truth_face**2)")
        else:
            #dr=dfq.eval(f"(theta_reco-theta_truth)*{z_front_face}")
            dx=dfq.eval("x_reco_face-x_truth_face"); dy=dfq.eval("y_reco_face-y_truth_face")
            dr=(dx*dfq.eval("x_truth_face")+dy*dfq.eval("y_truth_face"))/dfq.eval("sqrt(x_truth_face**2+y_truth_face**2)")
            
        a=dr
        print("sigma(dr)=", np.std(a))
        #if abs:
        #    a*=np.sign(x)
        a=a[(a>np.mean(a)-3*np.std(a))&(a<np.mean(a)+3*np.std(a))]
        
        bcs.append((energy_bins[j]+energy_bins[j+1])/2)
        sigma,dsigma,mu, dmu=res_calc(a, nsigma=2, quantile=quantile)
        sigmas.append(sigma)
        dsigmas.append(dsigma)
        mus.append(mu)
        dmus.append(dmu)
    
    plt.errorbar(np.array(bcs)*(1+shift), sigmas, dsigmas,label=label, marker='o', ls='', **arg)
    print(f"Energies={bcs} #GeV")
    print(f"sigma_thetas={list(np.array(sigmas)/z_front_face)} #mrad")
    print(f"dsigma_thetas={list(np.array(dsigmas)/z_front_face)} #mrad")
  
    
    
    plt.xlabel("Energy [GeV]")
    if direction=="phi":
        plt.ylabel("$\\sigma[r(\\phi_{\\rm recon}-\\phi_{\\rm truth})]$ [mm]")
    else:
        plt.ylabel("$\sigma_r$ [mm]")
    plt.ylim(0)
    return bcs, sigmas,dsigmas, mus, dmus
    
def yr_req(label='YR requirement: $\\frac{3\,\, mrad}{\\sqrt{E}}$',
          **arg):    
    #YR requirement
    x=np.linspace(10, 300,100)
    y=3/np.sqrt(x)*z_front_face
    plt.errorbar(x,y, label=label, **arg, marker='')

def write_geom(name="Fe",y=0.5,x=10, **arg):
    txt={"Fe":"Fe 20 mm\nSc 3.0 mm\n(64 layers)\ncell area 25 cm$^2$",
        "Pb":"Pb 10 mm\nSc 2.5 mm\n(110 layers)"}[name]
    plt.text(x, y, txt, fontsize=18,**arg)

def theta_axis():    
    #plt.legend()
    
    ax1=plt.gca()
    
    ax1.set_ylabel("$\sigma_r$ [mm]")
    
    ax2=plt.gca().twinx()
    ax2.set_ylim(ax1.get_ylim()[0]/z_front_face,ax1.get_ylim()[1]/z_front_face)
    
    ax2.set_ylabel("$\\sigma_\\theta$ [mrad]")
    return ax2

def fit_res(bcs, sigmas,dsigmas):
    P0=[10,2]
    fnc = lambda E, b,c: np.sqrt(b**2/E+c**2)
    P, cov =curve_fit(fnc,bcs, sigmas, p0=P0, sigma=dsigmas)
    x=np.linspace(8, 300,100)
    label="(%.2f\\pm %.2f)/\\sqrt{E}\\oplus\,(%.2f\\pm %.2f)$ mrad"%\
                                        (P[0]/z_front_face, np.sqrt(cov[0][0])/z_front_face, 
                                        P[1]/z_front_face, np.sqrt(cov[1][1])/z_front_face)
    
    print(label)
    label=f"fit: ${P[0]/z_front_face:.2f}/\\sqrt{{E}}\\,\\oplus\\,{P[1]/z_front_face:.2f}$ mrad"
                                        
    plt.errorbar(x, fnc(x,*P), label=label, color='purple', marker='',ls=':')

res_plot(df_h1, label="unstaggered", rw="", w0_params=(5.9,0.33,0.0), fit=False, color='tab:blue', fillstyle='none', shift=0)

res_plot(df_h4, label="staggered, baseline", rw="", w0_params=(4.2,0.045,0.32), fit=False, color='tab:orange', fillstyle='none', shift=0.007*3)
res_plot(df_h4, label="staggered, HEXPLIT", color='tab:green', fillstyle='full', shift=0.007*4)


#Bishnu's numbers
##------------THETA REGRESSION PARAMETERS---------------------------------------
##--Detector = zdc,   Absorber =64Fe,  input = 64Zseg,  output = 3D,     Model=graphnet,  ECAL Status=False
gnn_energy_Input64Zseg_output3D_zdc_64Fe=np. array([ 10.,  11.,  12.,  13.,  14.,  16.,  17.,  19.,  21.,  24.,  26.,
        29.,  32.,  35.,  39.,  43.,  47.,  52.,  57.,  63.,  70.,  77.,
        85.,  93., 103., 114., 125., 138., 152., 168., 185., 204., 225.,
       248., 274.])
gnn_theta_res_fit_Input64Zseg_output3D_zdc_64Fe=np. array([0.214664, 0.207316, 0.197218, 0.185306, 0.175127, 0.16065 ,
       0.156735, 0.151047, 0.143546, 0.134766, 0.125774, 0.117181,
       0.115651, 0.107514, 0.104951, 0.097164, 0.098249, 0.09131 ,
       0.085207, 0.085866, 0.076467, 0.073615, 0.072934, 0.070036,
       0.066807, 0.064527, 0.059164, 0.058777, 0.057299, 0.054317,
       0.055537, 0.05096 , 0.05005 , 0.049701, 0.049169])


plt.errorbar(gnn_energy_Input64Zseg_output3D_zdc_64Fe, gnn_theta_res_fit_Input64Zseg_output3D_zdc_64Fe*z_front_face,
            color='r', label="staggered, Graphnet", marker='s', ls='')


fit_res(gnn_energy_Input64Zseg_output3D_zdc_64Fe, gnn_theta_res_fit_Input64Zseg_output3D_zdc_64Fe*z_front_face, None)
yr_req(ls='--', color='tab:brown')

##------------PHI REGRESSION PARAMETERS---------------------------------------
##--Detector = zdc,   Absorber =64Fe,  input = 64Zseg,  output = 3D,     Model=graphnet,  ECAL Status=False
gnn_energy_Input64Zseg_output3D_zdc_64Fe=np. array([ 10.,  11.,  12.,  13.,  14.,  16.,  17.,  19.,  21.,  24.,  26.,
        29.,  32.,  35.,  39.,  43.,  47.,  52.,  57.,  63.,  70.,  77.,
        85.,  93., 103., 114., 125., 138., 152., 168., 185., 204., 225.,
       248., 274.])
gnn_phi_res_fit_Input64Zseg_output3D_zdc_64Fe=np. array([108.703249,  98.202715,  90.479059,  86.871533,  86.017905,
        81.101915,  77.404621,  71.527623,  68.538741,  65.895301,
        63.010289,  54.233386,  53.107992,  52.745853,  50.922052,
        47.958835,  48.982467,  42.819087,  42.817568,  41.003805,
        39.418247,  37.780723,  33.620471,  33.863215,  32.11117 ,
        30.793427,  30.856167,  28.776615,  29.274958,  28.067027,
        26.49414 ,  27.430675,  25.156428,  26.014058,  25.492691])

legend = plt.legend(frameon=False, facecolor='white', edgecolor='white', loc='upper right', fontsize=18)
legend.set_zorder(5)
#axs.yaxis.set_minor_locator(mpl.ticker.AutoMinorLocator(10))

# frame = legend.get_frame()
# frame.set_facecolor('white')
# frame.set_edgecolor('white')
plt.ylim(0, 20)
write_geom("Fe",15, 70)

theta_axis()

#plt.gca().set_xscale('log')
plt.gca().set_title("ZDC neutron simulation")

#frame.set_zorder(20)

plt.savefig("images/position_resolution_zdc_design_paper.pdf")
plt.show()
```
