```python
def row(X,Y):
    offsetX=0
    offsetY=0
    sidelength=1
    a=np.mod((Y-offsetY)/(np.sqrt(3)*sidelength)-.5,1)
    b=np.mod((X-offsetX)/(3/2*sidelength)-.5,1)
    return np.floor((X-offsetX)/(3*sidelength/2)+abs(b+a)-abs(a-b))
```

```python
import numpy as np
from matplotlib import pyplot as plt 
import matplotlib.cm as cm 
n = 256 
x = np.linspace(-3., 3., n) 
y = np.linspace(-3., 3., n) 
X, Y = np.meshgrid(x, y) 

Z = row(X,Y)

plt.pcolormesh(X, Y, Z, cmap = cm.gray) 
plt.show()
```
