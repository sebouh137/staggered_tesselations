```python

import numpy as np, pandas as pd

import matplotlib.pylab as plt
plt.rcParams["figure.figsize"] = [7.00, 7.00]
```

```python
def rap(E, pz=None, m=None, theta=None):
    if pz==None:
        pz=np.sqrt(E**2-m**2)*np.cos(theta)
        
    return 1/2*np.log((E+pz)/(E-pz)) 
```

```python
def pseudo(theta):
    return -np.log(np.tan(theta/2))
def angle(eta):
    return 2*np.arctan(np.exp(-eta))
```

```python
m=0.938
for eta in 3.0, 3.2, 3.4, 3.6, 3.8, 4.0:
    theta=angle(eta)
    E=np.linspace(0, 300, 300)
    y=rap(E, m=m, theta=theta)
    plt.plot(E,y,label=f"$\eta={eta:.1f}")
plt.xscale('log')
```
